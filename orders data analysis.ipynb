```python
#extracting file from zip file 
import zipfile 
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()  #extracting file to dir
zip_ref.close() 
```

```python
#reading data from the file
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#renaming columns in lower case and replacing space with underscore
#df.rename(columns={'Order Id' : 'order_id', 'City':'city'})
df.columns = df.columns.str.lower()
df.columns =df.columns.str.replace(' ','_')
df.head(5)
```

```python
#derive new columns discount in rs. for example- 2% of 260 is 5.2

df['discount']=df['list_price']*df['discount_percent']*.01
df.dtypes
```

```python
#new columns -> sale price = list price - discount 
df['sale_price']= df['list_price']-df['discount']
df
```

```python
# profit = sp - cp
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

```python
#conversion of order date from data type to date
df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df
```

```python
#drop cost, list and discount per columns 
df.drop(columns=['cost_price','list_price','discount_percent'])
```
